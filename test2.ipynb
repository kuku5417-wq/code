```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# 샘플 데이터셋 로드
tips = sns.load_dataset('tips')

# 산점도 생성
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time')
plt.title('Total Bill vs Tip')
plt.show()
```
